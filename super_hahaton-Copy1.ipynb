```python
#оптимизируем данные
full_data = data.drop(columns='floor').dropna()

full_data.columns
full_data['floor'] = data['floor']
```

```python
data.isna().sum()
data.head()

data['floor'] = data['floor'].astype('str')

# Функция из Nan в 1 этаж
def fromNaN(x):
    if x == 'nan' or x == '1.0':
        x = '1'
    elif x == 'цоколь' or x == 'подвал' or x =='цоколь':
        x = '-1'
    elif x == '-1.0':
        x = '-1'
    return x 

# Функция оптимизации))
def checkseq(y):
    if 'цоколь'  in y:
        y = '-1'
    elif 'подвал'  in y:
        y = '-1'
    elif 'Подвал'  in y:
        y = '-1'
    elif 'Цоколь'  in y:
        y = '-1'
    
    return y 

# Используем фуннкцию fromNan для столбца floor
data['floor'] = data['floor'].map(fromNaN)
data['floor'] = data['floor'].map(checkseq)

```

```python

data['floor'].replace(0, -1, inplace=True)  #заменил нулевые этажи на -1 , так как в данных есть как нулевые этажи, так и -1, хотя это одно и то же как по мне
```

```python
# Удаляем строки, в которых есть слова
data = data.drop(data[(data['floor'].map(len) > 5)].index) # алгоритм работает после перезагрузки дф
```

```python
# Удаляем строки, запись которых имеет вид  '4, 6 ', '1-й' и тд.
data = data.drop(data[(data['floor'] == '1, 2')].index)
data['floor'].unique()

```
