### Сводная таблица по кол-ву детей и отмен броней.

```python

number_of_childrenn = 0
for i in df['total_kids']:
    if  i == True :
        number_of_childrenn += 1  # Общее кол-во бронирований с детьми 
print(number_of_childrenn )
```

 ### Проведем расчет churn rate ( коэффициента оттока клиентов). Посмотрим как влияет отмена бронирования в зависимости от того, есть дети или нет.

```python
churn_without = round((df.query("has_kids == False & is_canceled == 1").shape[0]
       / df.query("has_kids == False").shape[0])*100,2) # Отток клиентов без детей
```

```python
churn_with = round((df.query("has_kids == True & is_canceled == 1").shape[0]
       / df.query("has_kids == True").shape[0])*100,2) # Отток клиентов с детьми

churn_without,\
churn_with
```

### Как мы видим из расчетов, отток клиентов с детьми меньше , чем отток клиентов без детьми 

```python
round(churn_without - churn_with, 2 )
```

#### Разница составляет 2,27 %
